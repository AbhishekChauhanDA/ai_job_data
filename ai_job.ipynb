```python
df.columns = df.columns.str.strip().str.lower()
string_cols = df.select_dtypes(include='object').columns
df[string_cols] = df[string_cols].apply(lambda x: x.str.strip())
```

appling for cleaning process {use for removing empty space in cells
}

```python
df['posting_date'] = pd.to_datetime(df['posting_date'], errors='coerce')
df['application_deadline'] = pd.to_datetime(df['application_deadline'], errors='coerce')
```

```python
plt.figure(figsize=(8,7))
sns.countplot(data=df, x='experience_level', order=df['experience_level'].value_counts().index)
plt.title('Experience Level Distribution')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(data=df['salary_usd'],kde=True,bins=40,color='skyblue')
plt.title('Salary Distribution (USD)')
plt.xlabel('Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='remote_ratio')
plt.title('Remote Work Distribution')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='experience_level', y='salary_usd', estimator='mean')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
top_jobs = df['job_title'].value_counts().head(10)
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Top 10 AI Job Titles')
plt.xlabel('Number of Jobs')
plt.ylabel('Job Title')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_countries = df['company_location'].value_counts().head(10)
sns.barplot(y=top_countries.index, x=top_countries.values)
plt.title('Top 10 Company Locations')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```
